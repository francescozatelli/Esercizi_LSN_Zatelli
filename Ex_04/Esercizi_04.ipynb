## 04.1 - Termalizzazione

Per implementare l'equilibrazione ho considerato delle simulazioni di 1000 step ciascuna, ripartendo ogni volta dalle ultime due configurazioni raggiunte. Una volta raggiunta una situazione stazionaria, ho salvato le ultime due configurazioni per ogni fase per poi utilizzarle nella simulazione vera e propria. 
Dai grafici sottostanti si nota innanzitutto che le fluttuazioni dell'energia totale (in ognuna delle simulazioni da 1000 step) sono molto più piccole rispetto a quelle dell'energia cinetica e dell'energia potenziale. Questo indica che l'algoritmo di Verlet, utilizzato per il moto delle particelle, conserva effettivamente l'energia localmente (la conservazione globale dell'energia invece non si può osservare nei grafici sottostanti, dato che le simulazioni sono corte), come deve essere dato che nell'insieme canonico sono fissati il numero di particelle $N$, il volume $V$ e l'energia $E$. La temperatura "target" è rappresentata con una linea nera tratteggiata.

### Solido

Per la termalizzazione della fase solida ho considerato 8 ripetizioni da 1000 step ciascuna. Si può notare che già circa dopo 4000 passi il sistema aveva raggiunto l'equilibrio. In un solido la componente cinetica dell'energia totale è piccola rispetto alla componente potenziale. Pertanto, una volta raggiunta l'equilibrazione, facendo ripartire il sistema più volte l'energia totale non cambia molto (si osservano infatti dei piccoli step tra una simulazione da 1000 passi e la successiva).

```python
import numpy as np
import matplotlib
import matplotlib.pyplot as plt

epot = np.loadtxt("Ex_04_1/Solid/output_epot.dat")
ekin = np.loadtxt("Ex_04_1/Solid/output_ekin.dat")
etot = np.loadtxt("Ex_04_1/Solid/output_etot.dat")
temp = np.loadtxt("Ex_04_1/Solid/output_temp.dat")
steps = 10*np.arange(len(temp)) #Una misura ogni dieci step

plt.rc('text', usetex=True)
plt.rc('font', family='serif')


fig, ax = plt.subplots(2, 2, figsize=(10,10),dpi=120)
ax[0,0].plot(steps, epot)
ax[0,0].set_title("Potential energy")
ax[0,0].set_ylabel(r"Energy $[\varepsilon]$")
ax[0,0].set_xlabel("Steps")
ax[0,0].grid(True)

ax[0,1].plot(steps, ekin)
ax[0,1].set_title(r"Kinetic energy")
ax[0,1].set_ylabel(r"Energy $[\varepsilon]$")
ax[0,1].set_xlabel("Steps")
ax[0,1].grid(True)

ax[1,0].plot(steps, etot)
ax[1,0].set_title(r"Total energy")
ax[1,0].set_ylabel(r"Energy $[\varepsilon]$")
ax[1,0].set_xlabel("Steps")
ax[1,0].grid(True)

ax[1,1].axhline(0.8,0,1,color="black", linestyle="dashed")
ax[1,1].plot(steps, temp)
ax[1,1].set_title(r"Temperature")
ax[1,1].set_ylabel(r"Temperature $[\varepsilon/k_B]$")
ax[1,1].set_xlabel("Steps")
ax[1,1].grid(True)
plt.show()
```

## 04.2/04.3 - Medie a blocchi e simulazioni

Per le simulazioni ho considerato $2\times 10^5$ step, effettuando una misura ogni $20$ step. Le $10000$ misure ottenute sono state quindi divise in $100$ blocchi da $100$ step ciascuno. Ogni simulazione parte dalle ultime due configurazioni ottenute dal punto precedente.

La linea nera tratteggiata rappresenta ancora la temperatura target, ma rappresenta soltanto un riferimento e non il valore medio né quello a cui dovrebbe convergere il valore della temperatura. Infatti, nell'insieme microcanonico (quello che stiamo considerando) non viene fissata la temperatura, bensì l'energia totale (oltre a numero di particelle e volume). Nel grafico della temperatura della fase gassosa ci sono dei valori che non sarebbero compatibili con il valore target, ma è sbagliato confrontarli con quest'ultimo.

In tutti i casi si può notare che l'incertezza finale sull'energia totale è molto minore rispetto a quella sull'energia cinetica e potenziale, a testimonianza del fatto che le fluttuazioni sulla prima sono più piccole anche globalmente.


Per l'argon, le unità di Lennard-Jones sono
 * $\sigma=0.34 \; \text{nm}$, unità di lunghezza
 * $\epsilon/k_B=120 \; \text{K}$, unità di temperatura
 * $m=39.948 \; \text{amu}$, unità di massa

Per ottenere i valori di energia e temperatura sopra riportati in unità del SI è sufficiente moltiplicare le energie per $ k_B\times120 \; \text{K}$ e le temperature per $120 \; \text{K}$. Riporto direttamente i valori nelle unità SI (le ottengo moltiplicando i valori nello script Python, i dati "grezzi" sono in unità di Lennard Jones).

Si può notare che la componente dell'energia potenziale è preponderante nel caso solido rispetto all'energia cinetica. Nella fase gassosa, invece è più importante il contributo dell'energia cinetica.

### Solido

```python
import scipy.constants as sc

nblocks, epot, err_epot = np.loadtxt("Ex_04_2/Solid/ave_epot.out", delimiter=",", usecols=(0,1,2), dtype="int,float,float", unpack=True)
nblocks, ekin, err_ekin = np.loadtxt("Ex_04_2/Solid/ave_ekin.out", delimiter=",", usecols=(0,1,2), dtype="int,float,float", unpack=True)
nblocks, etot, err_etot = np.loadtxt("Ex_04_2/Solid/ave_etot.out", delimiter=",", usecols=(0,1,2), dtype="int,float,float", unpack=True)
nblocks, temp, err_temp = np.loadtxt("Ex_04_2/Solid/ave_temp.out", delimiter=",", usecols=(0,1,2), dtype="int,float,float", unpack=True)

kb = sc.Boltzmann

plt.rc('text', usetex=True)
plt.rc('font', family='serif')


fig, ax = plt.subplots(2, 2, figsize=(12,12),dpi=120)
ax[0,0].errorbar(nblocks, epot*120*kb, yerr=err_epot*120*kb)
ax[0,0].set_title("Potential energy")
ax[0,0].set_ylabel("Energy [J]")
ax[0,0].set_xlabel("Number of blocks")
ax[0,0].grid(True)

ax[0,1].errorbar(nblocks, ekin*120*kb, yerr=err_ekin*120*kb)
ax[0,1].set_title("Kinetic energy")
ax[0,1].set_ylabel("Energy $[J]$")
ax[0,1].set_xlabel("Number of blocks")
ax[0,1].grid(True)

ax[1,0].errorbar(nblocks, etot*120*kb, yerr=err_etot*120*kb)
ax[1,0].set_title("Total energy")
ax[1,0].set_ylabel("Energy [J]")
ax[1,0].set_xlabel("Number of blocks")
ax[1,0].grid(True)

ax[1,1].axhline(0.8*120,0,1,color="black", linestyle="dashed")
ax[1,1].errorbar(nblocks, temp*120, yerr=err_temp*120)
ax[1,1].set_title("Temperature")
ax[1,1].set_ylabel("Temperature [K]")
ax[1,1].set_xlabel("Number of blocks")
ax[1,1].grid(True)
plt.show()
```

I valori finali ottenuti sono, in unità di Lennard-Jones e naturali:
$$E_{pot}=-6.6612 \pm 0.0004 \; \varepsilon= (-1.10362 \pm 0.00007)  \times 10^{-20}  \; \text{J}$$
$$E_{kin}=1.2002 \pm 0.0004 \; \varepsilon = (1.9884 \pm 0.0007)\times 10^{-21} \; \text{J}$$
$$E_{tot}=-5.46104 \pm 0.00007 \; \varepsilon = (-9.04773 \pm 0.0011)\times 10^{-21} \; \text{J}$$
$$T = 0.8001 \pm 0.0003 \; \frac{\varepsilon}{k_B}= 96.02 \pm 0.03 \; \text{K}$$

## 04.4 - Pressione

Ho infine calcolato la pressione sfruttando il teorema del viriale. Le simulazioni sono analoghe a quelle discusse nel punto precedente. La pressione, in unità di Lennard-Jones è espressa in $\varepsilon/\sigma^3$. I grafici seguenti sono già in unità SI (per l'argon).

```python
sigma=0.34e-09
epsilon=120*kb

nblocks, p_solid, err_p_solid = np.loadtxt("Ex_04_4/Solid/ave_pres.out", delimiter=",", usecols=(0,1,2), dtype="int,float,float", unpack=True)
nblocks, p_liquid, err_p_liquid = np.loadtxt("Ex_04_4/Liquid/ave_pres.out", delimiter=",", usecols=(0,1,2), dtype="int,float,float", unpack=True)
nblocks, p_gas, err_p_gas = np.loadtxt("Ex_04_4/Gas/ave_pres.out", delimiter=",", usecols=(0,1,2), dtype="int,float,float", unpack=True)

plt.rc('text', usetex=True)
plt.rc('font', family='serif')


fig, ax = plt.subplots(3, 1, figsize=(8,13))
fig.tight_layout(pad=3.0)
ax[0].errorbar(nblocks, p_solid*epsilon/sigma**3, yerr=err_p_solid*epsilon/sigma**3)
ax[0].set_title("Solid - Pressure")
ax[1].errorbar(nblocks, p_liquid*epsilon/sigma**3, yerr=err_p_liquid*epsilon/sigma**3, color="C1")
ax[1].set_title("Liquid - Pressure")
ax[2].errorbar(nblocks, p_gas*epsilon/sigma**3, yerr=err_p_gas*epsilon/sigma**3, color="C2")
ax[2].set_title("Gas - Pressure")

for axi in ax:
    axi.set_ylabel("Pressure [Pa]")
    axi.set_xlabel("Number of blocks")
    axi.grid(True)
plt.show()
```

Si può notare che la pressione è massima per la fase solida e minima per la fase gassosa, come era ragionevole attendersi. I valori finali sono:
$$P_{sol}= 9.607 \pm 0.003 \; \frac{\epsilon}{\sigma^3} = (4.0498 \pm 0.0010)\times 10^8 \; \text{Pa}$$
$$P_{liq}= 2.166 \pm 0.013 \; \frac{\epsilon}{\sigma^3} = (9.13 \pm 0.05)\times 10^7 \; \text{Pa}$$
$$P_{gas}= 0.0494 \pm 0.0002 \; \frac{\epsilon}{\sigma^3} = (2.084 \pm 0.009)\times 10^6 \; \text{Pa}$$
